```python
import pandas as pd
from helper_functions import data_preprocess, heat_map_plot
```

```python
sl1 = pd.read_parquet('sidelink1_df_v2.parquet', engine='pyarrow')
sl2 = pd.read_parquet('sidelink2_df_v2.parquet', engine='pyarrow')
sl3 = pd.read_parquet('sidelink3_df_v2.parquet', engine='pyarrow')
location = pd.read_parquet('localization_data_v2.parquet', engine='pyarrow')
measurement_plan = pd.read_parquet('meas_plan.parquet', engine='pyarrow')
```

```python
measurement_plan
```

```python
Dict = dict()
time_interval = 5 # in minutes
cntr1 = 0
cntr2 = 0
for start_time, end_time, wall_scenario in zip(measurement_plan['start_time'], measurement_plan['end_time'], measurement_plan['wall_scenario']):        
    for sl in list(measurement_plan.columns)[-3:]:
        if measurement_plan[sl][cntr1] == True:
            sl_ind = int(list(sl)[2])            
            print('Processing ' + str(start_time) + ' / ' + str(end_time) + ' wall scenario ' + wall_scenario + ' for sl_' + str(sl_ind))
            Dict['preprocessed_' + str(cntr2 + 1)] = data_preprocess(start_time, end_time, time_interval, globals()['sl' + str(sl_ind)], location, sl_ind, wall_scenario)            
            print('###############')
            cntr2 = cntr2 + 1
    cntr1 = cntr1 + 1
```

```python
merged = pd.concat(Dict.values(), ignore_index=True)
```
